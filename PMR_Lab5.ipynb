## **Box Plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 


#General random data for  the box plot 
np.random.seed(42)
data=[np.random.normal(50,10,100),np.random.normal(70,15,100),np.random.normal(60,5,100)]

#create the box plot 
plt.figure(figsize=(8,6))
sns.boxplot(data=data)
plt.xticks([0,1,2],['Group 1','Group 2','Group 3'])
plt.title('Box Plot')
plt.ylabel('values')
plt.show()
```
